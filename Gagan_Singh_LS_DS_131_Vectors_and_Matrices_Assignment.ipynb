## 1.2 Create a three-dimensional vecor and plot it on a graph

```
from mpl_toolkits.mplot3d import Axes3D

yellow = [1, 2, 3]
red = [3, 4, 5]
blue = [6, 7, 7]

vectors = np.array([[0, 0, 0, 1, 2, 3], 
                    [0, 0, 0, 3, 4, 5],
                    [0, 0, 0, 6, 7, 7]])
X, Y, Z, U, V, W = zip(*vectors)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
from math import e, pi
print(e)
print(pi)

#select a vector
yellow = [1,.5]

#multiply by scaler

blue = np.multiply(5, yellow)

red = np.multiply(math.pi, yellow)

orange = np.multiply(-e,yellow)

plt.arrow(0,0, blue[0], blue[1],head_width=.05, head_length=0.05, color ='blue')
plt.arrow(0,0, red[0], red[1],head_width=.05, head_length=0.05, color ='red')
plt.arrow(0,0, yellow[0], yellow[1],head_width=.05, head_length=0.05, color ='yellow')
plt.arrow(0,0, orange[0], orange[1],head_width=.05, head_length=0.05, color ='orange')
plt.xlim(-1,8)          
plt.ylim(-1,8)
plt.title("Scaled Vectors")
plt.show()
```

One thing I notice is that these vectors are all on the same line, in a diagonal trend moving upwards, whatever we scale the vector to will portray back on the graph.
